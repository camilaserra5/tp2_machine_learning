```python
import pandas as pd
```

```python
randomforest =  pd.read_csv('RandomForestResults.csv', low_memory=False)
xgboost=  pd.read_csv('resultado_xgboost.csv', low_memory=False)
adaboost =  pd.read_csv('resultados_adaboost.csv', low_memory=False)
```

```python
df = pd.merge(randomforest, xgboost,on='person',how='inner')
df = pd.merge(df,adaboost,on='person',how='inner')
```

```python
df
```

```python
resultados = []
for indice_fila, fila in df.iterrows():
    cantidad_de_1 = 0
    for x in range(3):#aca el numero de columnas
        if fila[x+1] == 1:
            cantidad_de_1+=1
    if cantidad_de_1>=2:#aca la mitad de las columnas
        resultados.append(1)
        continue
    resultados.append(0)
```

```python
resultado_final = pd.read_csv("trocafone_kaggle_test.csv", low_memory=False)
resultado_final['label'] =resultados
```

```python
resultado_final.to_csv('vpm.csv',index=False)
```
