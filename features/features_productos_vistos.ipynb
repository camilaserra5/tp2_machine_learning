### Notebook para la creacion de features sobre productos vistos, por ejemplo, la cantidad de diferentes prod vistos, la cant maxima de visitas al mismo producto, la ctd total de prod vistos. Al finalizar genera un archivo csv con nombre "features_productos_vistos.csv"

```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, timedelta
import utils
```

```python
events = utils.get_events()
events.head()
```

```python
df_viewed = events[events['event'] == 'viewed product']
df_viewed = df_viewed[['timestamp', 'person', 'sku']]
df_viewed.head()
```

```python
df_viewed = df_viewed.groupby(['person', 'sku']).agg({'timestamp':'count'})
df_viewed = df_viewed.reset_index()
df_viewed.head()
```

```python
df_viewed = df_viewed.groupby('person').agg({'sku':'count', 'timestamp':['sum','max']})
df_viewed = df_viewed.reset_index()
df_viewed.columns = ['person', 'different_products_views', 'total_views', 'same_product_max_views']
df_viewed.head()
```

```python
df_viewed.to_csv('features_productos_vistos.csv',index=False)
```
