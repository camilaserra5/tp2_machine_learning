```python
# imports iniciales
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, timedelta
import utils
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)
```

```python
events = pd.read_csv('events_up_to_01062018.csv',low_memory=False)
events.head()
```

```python
feature = events[['person','timestamp']]
feature.head()
```

```python
def devolver_int(fecha):
    numero = ''
    for x in fecha:
        if x.isdigit():
            numero = numero + x
    return numero

# feature['fecha_int'] = events['timestamp']
feature['fecha_int'] = feature['timestamp'].map(lambda x: int(devolver_int(str(x))))
feature.head()
```

```python
feature = feature.groupby('person')['fecha_int'].agg({'suma':sum,'promedio':'mean','maximo':max,'minimo':min})
feature.reset_index(inplace=True)
```

```python
feature.to_csv('features_fechas_int.csv',index=False)
```
