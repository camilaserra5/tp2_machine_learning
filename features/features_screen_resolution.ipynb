```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, timedelta
import utils
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)
```

```python
events = pd.read_csv('events_up_to_01062018.csv',low_memory=False)
events.head()
```

```python
screen_resolution = events[['person','screen_resolution']]
screen_resolution['screen_resolution'].fillna(0,inplace=True)
screen_resolution.head()
```

```python
screen_resolution['screen_resolution'].value_counts()
```

```python
def calculo(screen):
    if screen == "0":
        return 0
    screen_  = screen.split('x')
#     print(int(screen_[0]))
    return int(screen_[0])*int(screen_[1])

screen_resolution['calculos_resolution'] = screen_resolution['screen_resolution'].map(lambda x: int(calculo(str(x))))
screen_resolution = screen_resolution[['person','calculos_resolution']]
screen_resolution.head()
```

```python
screen_resolution = screen_resolution.groupby('person')['calculos_resolution'].agg({'suma_resolution':sum,'promedio_resolution':'mean','maximo_resolution':max,'minimo_resolution':min})
screen_resolution.reset_index(inplace=True)
screen_resolution.head()
```

```python
screen_resolution.to_csv('features_screen_resolution.csv',index=False)
```
