```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, timedelta
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)
```

```python
eventos1 = pd.read_csv('events_up_to_01062018.csv',low_memory=False)
eventos2 = pd.read_csv('events.csv',low_memory=False)
```

```python
eventos1.head()
```

```python
eventos2.head()
```

```python
print(eventos1.shape)
print(eventos2.shape)
```

```python
eventos = eventos1.append(eventos2, ignore_index = True)
eventos.shape
```

#### Me quedo con las compras y le agrego label 

```python
compras = eventos[eventos['event'] == 'conversion']
compras.shape
```

```python
compras['label'] = 1
```

```python
compras = compras[['person','label']]
compras.head()
```

#### Vuevlo a juntar y escribo csv

```python
eventos = pd.merge(eventos, compras, on = 'person',how = 'left')
eventos['label'] = eventos['label'].map(lambda x: 1 if x==1 else 0)
eventos.head()
```

```python
eventos['label'].value_counts()
```

```python
eventos.to_csv('../modelos/set_entrenamiento.csv',index=False)
```

```python
training = eventos[['person','label']]
training = training.drop_duplicates(subset = 'person')
set_reducido_0 = training[training['label'] == 0].sample(n=10000)
set_entrenamiento = training[training['label'] == 1].append(set_reducido_0)
set_entrenamiento.shape
set_entrenamiento.to_csv('../modelos/labels_training_set.csv',index=False)
```
